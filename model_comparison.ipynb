#  Final Comparison of All Models (CNN, LSTM, Bi-LSTM, BERT, RoBERTa, DistilBERT)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cnn_lstm_df = pd.read_csv("metrics/cnn_lstm_metrics2.csv")

# Rename columns to match transformer format
cnn_lstm_df = cnn_lstm_df.rename(columns={
    "model": "Model",
    "accuracy": "Accuracy",
    "f1": "F1-Score",
    "params": "Parameters"
})
```

```python
transformer_df = pd.read_csv("metrics/transformer_model_comparison.csv")

# If 'Parameters' not included, add dummy column
if "Parameters" not in transformer_df.columns:
    transformer_df["Parameters"] = None
```

```python
metrics_df = pd.concat([cnn_lstm_df, transformer_df], ignore_index=True)
metrics_df.to_csv("model_comparison_results.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for metric in ["Accuracy", "F1-Score", "Precision", "Recall"]:
    plt.figure(figsize=(10, 5))
    sns.barplot(data=metrics_df, x="Model", y=metric, palette="Set2")
    plt.title(f"Model {metric} Comparison")
    plt.ylim(0.3, 1.0)
    plt.ylabel(metric)
    plt.xticks(rotation=15)
    plt.tight_layout()
    plt.show()
```
